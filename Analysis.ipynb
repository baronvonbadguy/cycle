This function just sets us up to be able to render heatmaps of values.  It's a way to represent a number range as a color range, which is great for proportional or relative values.

```python
#borrowed this from https://gist.github.com/gjx/6661288
def heatmap(df,
            edgecolors='w',
            cmap=mpl.cm.get_cmap('Spectral'),
            log=False,
            scale=4):    
    width = len(df.columns)/scale
    height = len(df.index)/scale
    fig, ax = plt.subplots(figsize=(width,height))
    heatmap = ax.pcolor(df,
                        edgecolors=edgecolors,  # put white lines between squares in heatmap
                        cmap=cmap,
                        norm=mpl.colors.LogNorm() if log else mpl.colors.Normalize(vmax=.999, vmin=0, clip=True))
    
    ax.autoscale(tight=True)  # get rid of whitespace in margins of heatmap
    ax.set_aspect('equal')  # ensure heatmap cells are square
    ax.xaxis.set_ticks_position('bottom')  # put column labels at the top
    ax.tick_params(bottom='off', top='off', left='off', right='off')  # turn off ticks
    
    plt.yticks(np.arange(len(df.index)) + 0.5, df.index)
    plt.xticks(np.arange(len(df.columns)) + 0.5, df.columns, rotation=90)
    
    # ugliness from http://matplotlib.org/users/tight_layout_guide.html
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", "3%", pad="1%")
    plt.colorbar(heatmap, cax=cax)
```

```python
new_format_sheets = [sheet for sheet in workbook.sheet_names() if re.search('201[0-2]', sheet)]
old_format_sheets = [sheet for sheet in workbook.sheet_names() if re.search('200[0-9]', sheet)]
actuals_sheets = set(workbook.sheet_names()) - (set(new_format_sheets)|set(old_format_sheets))
actuals_sheets
```

Right here we're going to define the 3 grids in which our values are going to lie.  We're going to build this by permutating the material types with their respective market locations.  We do this three times, each grid with a different offset marking the top-left origin.

```python
from collections import defaultdict
headers_stock_col_1 = ['Mixed Paper (1)', 'Boxboard Cuttings (4)', 
                   'News (6)', 'News (8)', 'OCC (11)', 'SOP (37)', 'Sorted White Ledger (40)']
headers_stock_index_1 = ['Chicago', 'New England', 'Buffalo', 'New York', 
                         'Southeast', 'Southwest', 'Los Angeles', 'San Francisco', 'Pacific NW']
headers_stock_col_2 = ['New DLK Cuttings (13)', 'White News Blanks (24)', 'Coated Soft White Shavings(28)', 
                       'Hard White Shavings (30)', 'Hard White Env. Cuts(31)', 'Unprint Bleached Sulfate (47)']
headers_stock_index_2 = ['East', 'Midwest', 'Southeast', 'Southwest', 'Los Angeles', 'San Francisco', 'Pacific NW']
headers_export_col = ['Mixed Paper (1)', 'News (8)', 'OCC (11)']
headers_export_index = ['Chicago', 'New York', 'Los Angeles']
headers = defaultdict(dict)
for x, col in enumerate(headers_stock_col_1):
    for y, index in enumerate(headers_stock_index_1):
        headers[x + 1][y + 7] = (col, index)
for x, col in enumerate(headers_stock_col_2):
    for y, index in enumerate(headers_stock_index_2):
        headers[x + 1][y + 21] = (col, index)
for x, col in enumerate(headers_export_col):
    for y, index in enumerate(headers_export_index):
        headers[x + 1][y + 33] = (col, index)
```

This code is what actualy retrieves us the information for the old format.  It loops through every cell of every sheet and checks the location of that cell with the locations of known cells we want.  If the location is found in the headers dict keys, then it takes the values of that location in the sheet and pairs it with the values in the headers dict designating the name of the cell.  This is the most efficient way I could think of to grab all of our data with the proper names associated with each cell.

```python
df = pd.DataFrame()
for i, sheet_name in enumerate(old_format_sheets[2:]):
    sheet = workbook.sheet_by_name(sheet_name)
    db = dict()
    for col in range(sheet.ncols):
        for row in range(sheet.nrows):
            try:
                if col in headers.keys():
                    if row in headers[col].keys():
                        paper_type = headers[col][row][0]
                        paper_market = headers[col][row][1]
                        value = sheet.cell_value(row, col)
                        low = re.search('(\d+)-', value)
                        hi = re.search('-(\d+)', value)
                        if low:
                            db[(paper_type, paper_market, 'low')] = low.group(1)
                        else:
                            db[(paper_type, paper_market, 'low')] = np.nan
                        if hi:
                            db[(paper_type, paper_market, 'hi')] = hi.group(1)
                        else:
                            db[(paper_type, paper_market, 'hi')] = np.nan
            except Exception as e:
                print(e)
    if i == 0:
        df = pd.DataFrame(db.values(), index=pd.MultiIndex.from_tuples(db.keys()))
        df.columns = [sheet_name]
    else:
        df[sheet_name] = db.values()
df = df.sort(ascending=True).T
```

Here we're more or less repeating the same process with different offsets, and while grabbing both the low and high counts, which are now separate cells.

This is a handy little function to take a number of date formats and convert it to one that can be read by the DatetimeIndex constructor.

```python
from datetime import datetime
def try_parsing_date(text):
    #manual tweak
    text = re.sub('Sept(ember)?', 'Sep', text)
    for fmt in ('%b %Y', '%b. %Y', '%B %Y', '%b%y', '%Y-%m-%e'):
        try:
            date = datetime.strptime(text, fmt)
            return date
        except ValueError:
            pass
```

Let's add some data from Wood Pulp futures from The World Bank, brought to us by Quandl.  The CPI index is to adjust for inflation rate and convert nominal dollars to real dollars.  The PPICRM is the Producer Price Index for Crude Materials for Further Processing.  While we're at it, let's also just grab a bunch of indeces that might help us track the nations import/export of certain materials.

```python
def get_quandl(dbs):
    headers = {'X-API-Token': 'nDpdzhVa3EcWL3RjEcE4'}
    dfs = list()
    for qid, cols in dbs:
        endpoint = 'https://www.quandl.com/api/v2/datasets/' 
        url = endpoint + qid + '/data'
        r = rq.get(url, headers=headers)
        if r.status_code == 200:
            rjson = r.json()
            df = pd.DataFrame(rjson['data'], columns=rjson['column_names'])
            df.set_index('Date', drop=True, inplace=True)
            df.set_index(pd.to_datetime(pd.Series(df.index)))
            if len(df.columns) == 1:
                df.columns = [qid.split('/')[-1]]
            if cols:
                match = list(set(rjson['column_names']) & set(cols))
                dfs.append(df[match])
            else:
                dfs.append(df)
    comp = pd.concat(dfs, axis=1)
    comp.index = pd.DatetimeIndex(comp.index)
    comp = comp.resample('MS', how='max')
    return comp

dbs = [('WORLDBANK/WLD_WOODPULP',[]), 
       ('FRED/CPILFESL',[]), 
       ('FRED/PPICRM',[]),
       ('USCENSUS/PROD_E_5700', ['(12420) Pulpwood and woodpulp',
                                 '(12430) Newsprint',
                                 '(21150) Pulp and paper machinery']),
       ('BLS/EIUIP48',[]), ('BLS/EIUIPX',[]), 
       ('BLS/EIUIDX',[]), ('BLS/EIUID48',[]), 
       ('BLS/EIUID47',[])]

dfq = get_quandl(dbs)
dfq[-10:]
```

Now we're going to get the velocity and acceleration transforms of our Avg. Price data, then we'll cross correlate each transform among paper types, and finally composite the two and create a heatmap to visualize.  We're going to use Pearson correlation on the Velocity values, because they are absolute values and increase linearly, and Spearman on the Acceleration because it is proportional to the magnitude of the velocity and does not increase linearly.  

[Spearman](http://simple.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient)
[Pearson](http://en.wikipedia.org/wiki/Pearson_product-moment_correlation_coefficient)
[How to Choose](http://stats.stackexchange.com/questions/8071/how-to-choose-between-pearson-and-spearman-correlation)

I did this so that we can study the proportion of movement, not the magnitude.  This is important because I'm not interested if Mixed (2) fluctuated 5 bucks on the same day that EXPORT-OCC (11) fluctuated 5 bucks if thats 3% of the value of OCC while it's 10% the value of Mixed.

```python
def get_vel_acc(df):
    '''Returns velocity and acceleration dataframes of input, 
    acceleration values in units of percent change'''
    vel = pd.rolling_apply(df, 2, lambda x: (x[1] - x[0]))
    acc = pd.rolling_apply(vel, 2, lambda x: (x[1] - x[0]) / x[0])
    return (vel, acc)

vel, acc = get_vel_acc(df_avg_all)
vcorr, acorr = vel.corr(method='pearson'), acc.corr(method='spearman')

def upper_lower(u, l):
    '''Returns composite dataframe of comination of upper 
    and lower halves of respective inputs'''
    u = vcorr.mask(np.triu(u.astype(bool)))
    l = acorr.mask(np.tril(l.astype(bool)))
    return u.fillna(0) + l.fillna(0)

comp = upper_lower(vcorr, acorr)
heatmap(comp, scale=2.5)
```

Perfect, we got some nibbles.  Here we have the Velocity differential correlations on the top half, and Acceleration on the bottom.  The redder colors represent low correlations, and the bluer colors represent higher correlations.  We're primarily looking at the lower acceleraiton half, because it's the rate of change in proportion to the magnitude of change, but the velocity is there in case we need to refer to it later.

```python
def highest_correlators(df):
    max_ids = df[df < 1].idxmax()
    max_vals = df[df < 1].max()
    maxes = pd.concat([max_ids, max_vals], axis=1, keys=['Highest Correlator', 'Value'])
    maxes = maxes.drop_duplicates(subset=['Value']).sort(columns=['Value'], ascending=False)
    return maxes
highest_correlators(acorr)
```
